<b>IMPORT NECESSARY LIBRARIES</b>

```python
import pandas as pd
import numpy as np
```

<B>READ THE FILES TO A DATAFRAME</B>

```python
df=pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
df.head()
```

<B>REMOVE THE UNIMPORTANT COLUMNS</B>

```python
#Remove the columns 'Corporate Author', 'Corporate Contributors',
#'Former owner','Engraver','Edition Statement','Issuance type','Shelfmarks'
columns_toDrop = ['Corporate Author', 'Corporate Contributors','Former owner','Engraver','Edition Statement','Issuance type','Shelfmarks']
df.drop(columns_toDrop, inplace=True, axis=1)
df.head()
```

<b>CHECK IF THE 'IDENTIFIER' COLUMN IS UNIQUE</b>

```python
df['Identifier'].is_unique
```

<b>MAKE THE HE 'IDENTIFIER' COLUMN AS INDEX</b>

```python
df = df.set_index('Identifier')
df.head()
```

<b>CLEANING THE DATE OF PUBLICATION COLUMN</b>

```python
# extracting values whose first 4 characters are digits
extr = df['Date of Publication'].str.extract('^(\d{4})',expand=False)
# Converting the extracted values to numeric and assign to data frame column
df['Date of Publication'] = pd.to_numeric(extr)
#print(extr)
```

<b>CLEANING PLACE OF PUBLICATION</b>

<b>Using np.where</b><br>
<pre>  Syntax<br>
</t>np.where(condition, then, else)
<BR>It can be nested

```python
# Replace the ''Place of Publication'' with London and Oxford if it contains that place.
df['Place of Publication']=np.where(df['Place of Publication'].str.contains('London'),'London',
                                   np.where(df['Place of Publication'].str.contains('Oxford'),'Oxford',
                                   df['Place of Publication'].str.replace('_',' ')))
```

```python
# Retrieving columns with 'Place of Publication' with Oxford
df[df['Place of Publication']=="Oxford"]
```

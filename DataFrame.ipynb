```python
#PANDAS
    # A popular library for data analyis
#DATAFRAMES
    # Dataframe is a two dimensional datastructures in pandas.
    # Data is arranged in the form of rows and columns.
```

```python
# Note: The dataset used here is the movie dataset from Kaggle.
# The datasource: movies_metadata.csv
# https://www.kaggle.com/rounakbanik/the-movies-dataset 
```

```python
import pandas as pd # Importing the pandas library
```

```python
#Load the data in the csv file to a dataframe using the built-in function read_csv
#df is the dataframe
df=pd.read_csv('movies_metadata.csv') #COPY FILE PATH HERE
df.head()#Displays the first 5 rows in the dataframe
```

```python
# Creating a new dataframe from the existing one.
new_df=df[['original_title','overview','genres','original_language','release_date']] # New dataframe has 5 colums
new_df.head(10)# Displays only the first 10 rows.
```

```python
#Accessing elements from dataframe
```

```python
new_df.iloc[1,0] # Accessing second row first column
```

```python
new_df.iloc[1,1]  # Accessing second row first column
```

```python
#Accessing column using column-name
new_df.loc[1,'original_title']
```

```python
#Slicing the dataframe
new_df.iloc[0:2,0:3] # Two rows and 3 columns
```

```python
#Slicing the dataframe using column names
new_df.loc[0:3,'original_title':'genres']
```

```python
#Getting column as a dataframe.
    #Use two square brackets
lan_df=new_df[['original_language']] 
lan_df.head() # lan_df is a dataframe
```

```python
#Getting column as a series.
    # Series is a one dimensional dataframe
    # Use just one bracket
x=new_df['original_language']
x.head()
```

```python
new_df['original_language'].unique()#Listing unique elements
```

```python
en_movie_df=new_df[new_df['original_language']=='en'] #Select only English movies and assign to a new dataframe 'en_movie'
en_movie_df
```

```python
#Saving the dataframe to a CSV file
en_movie_df.to_csv('Eng_movies.csv')#COPY THE FILE PATH
```
